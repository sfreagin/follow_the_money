### Scratch notes playing with aggregated data

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
aapl_df = pd.read_csv('SCT_datasets/AAPL_SCTpay.csv')
f_df = pd.read_csv('SCT_datasets/F_SCTpay.csv')
googl_df = pd.read_csv('SCT_datasets/GOOGL_SCTpay.csv')
intc_df = pd.read_csv('SCT_datasets/INTC_SCTpay.csv')
jnj_df = pd.read_csv('SCT_datasets/JNJ_SCTpay.csv')
kdp_df = pd.read_csv('SCT_datasets/KDP_SCTpay.csv')
kmi_df = pd.read_csv('SCT_datasets/KMI_SCTpay.csv')
msft_df = pd.read_csv('SCT_datasets/MSFT_SCTpay.csv')
swks_df = pd.read_csv('SCT_datasets/SWKS_SCTpay.csv')
vmw_df = pd.read_csv('SCT_datasets/VMW_SCTpay.csv')
```

```python
main_df = pd.concat([aapl_df,f_df,googl_df,intc_df,jnj_df,kdp_df,kmi_df,msft_df,swks_df,vmw_df],ignore_index=True).reset_index()
main_df.drop(columns='index', inplace=True)
```
