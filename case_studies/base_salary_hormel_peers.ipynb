## Base Salary and Peer Groups

Companies must pay employees for their work; this is self-evidently true for legal reasons, economic reasons, societal reasons, moral and ethical reasons, and others. *How much* a given employee should be paid is also a matter for much discussion. The risk for the employer is, if you underpay people then they are more likely to leave for a competitor who pays more. If you overpay people then you waste resources that could go towards new employees or new investments. That should be clear enough in the abstract, but what does it mean to overpay someone? What does it mean to underpay someone?

The answers depend on many factors but the general rule of thumb is: the more valuable an employee's contributions are, the higher should be their pay. 

> ---
> **_NOTE:_** *Value* is difficult to measure, and for a given employee we might use metrics like seniority, prior experience, service time, sales generated, performance goals, perceived talent, and so on. 
>
>Suffice to say these measurements are often vague, subject to personal biases, and/or flat-out irrelevant to measuring the employee's actual contributions.
>
> ---

Compensation professionals, including the providers of market compensation surveys, often review multiple factors to assess how much "the market" currently values a given range of jobs.  Some of the more common factors are listed below. 

**Role**: what is the actual job being performed? Is this a day trader, or a pharmacology scientist, or a legal advisor, or...? Some roles will *intrinsically* command higher value than others. As an arbitrary example: all else being equal, in today's marketplace an entry-level chemical engineer will likely demand higher wages than an entry-level accountant.

**Industry**: all industries are competitive, but different industries compete in different ways.  For example, a national retailer more often competes on *volume* and *size* whereas a cloud computing company more often competes on *technical expertise*. Some industries have lots of lower-paid entry level employees (*e.g.* fast food restaurants) whereas other industries have fewer highly-paid employees (*e.g.* pharmaceutical research). In newer, growing, or "hot" industries, companies are often willing to pay more to bring on required talent.

**Experience**: a person with more experience in a role will tend to be more valued than someone with less experience. For example: all else being equal, a person with 15 years of experience in Human Resources is expected to make more contributions than a person just entering the HR job market. And here, *more contributions* tends to mean *more valuable*.

**Geography**: companies and employees in different geographies are subject to differences in cost of living, labor availability, resources, business opportunities, and so on. This tends to create unequal leveling in wages across regions. For example: a manufacturing employee in the United States will likely demand higher wages than an equivalent manufacturing employee in the Philippines.

> **_NOTE:_** From here we might start combining criteria to understand how pay structures can rank against one another. For example:
>* A back-end kernel software engineer is more valuable to a cloud computing company than to a regional clothing retailer 
>* An expert in supply chain logistics is more valuable to an multinational beverage company than to an Artificial Intelligence startup company
>* An entry-level chemical engineer may be less valuable to an energy company than a Senior Director of Compliance with 20 years of experience
>* A graphic designer with 5 years of experience in India may or may not be more valuable to an advertising firm than a New York-based sales representative fresh out of college
>* A software company will value software engineers over financial experts
>    * A Wall Street bank will value financial experts over software engineers

This is a lot to consider for any single job, let alone the many thousands of jobs that require level-setting at a given company. And so the most common practice amongst compensation professionals is to *look at what everyone else is paying* and adjust accordingly. This is typically called **benchmarking** and it also represents the most defensible position, because if it doesn't work you can say, "we are paying according to the market data," whereas creating your on method requires lots of creative thought as to *why* these jobs are valuable, and you must be prepared to defend your philosophical position if things go wrong. Much easier to just copy everyone else's work.

When benchmarking, we tend to implictly assume that everyone else made good compensation decisions for us to copy. But more likely is, everyone else *also* just looked at everyone else to set levels. And so the whole field is self-generating, based on everyone looking at everyone else and adding their own minor adjustments here and there. The net result of this, I argue, is the *persistent ratcheting of pay increases* which are divorced from the actual value of jobs.

It's not immediately obvious that pay should increase as a result of looking at peers; indeed it seems just as likely for pay to remain stagnant, or even to enter a downward spiral. Therefore I will explore the issue on a smaller scale, looking at base salaries for a group of top executives and demonstrate how the mathematics will *naturally lead to an increase in base salaries*.


### Peer Groups
Companies use **peer groups** as a way to gauge whether or not their compensation practices are aligned with the general market for talent. The idea is, rather than looking at the *entire* market you should only look at those peers with whom you compete for talent. For example, a beverage company like Coca-Cola (KO) would consider Pepsico (PEP) and Keurig Dr Pepper (KDP) to be peers, but not Microsoft (MSFT) or Ford (F) since these companies are unlikely to compete in the beverage industry.

Peer groups are often created by or with the support of an external compensation consultant. Companies tend to be in similar industries, and may often be competitors for market talent. The size of each peer company, most often measured by annual Revenues or a similar financial metric, will most typically range something like 50% to 200% of the company's own size.



Peer groups are disclosed in the CD&A section of a company's DEF 14A Definitive Proxy statement, filed annually ahead of the company's required annual shareholder meeting. As an example, here is the peer group discussion that Post Holdings (POST) discloses on page 29 of their [fiscal 2021 annual proxy](https://www.sec.gov/Archives/edgar/data/1530950/000153095021000415/fy2021defproxystatement.htm):

>*Annually, the Committee reviews total compensation market data provided by Aon. The Committee reviews and approves the peer group used for comparisons prior to commencement of the pay study. Consistent with prior years, the following peer group development criteria were used to develop competitive market values to assist with fiscal year 2021 pay decisions:*
>
>* **Industry**: *similar to Post, based on the Global Industry Classification System (GICS) code of Packaged Foods & Meats;*
>* **Company size**: *approximately 0.4 times to 3 times Post’s annual revenues, with a secondary focus on market cap;*
>* **Peers**: *companies using Post in their compensation peer group;*
>* **Peers of peers**: *companies used in the peer groups of potential peer companies; and*
>* **Competitors**: *companies that compete with Post for business and management talent.*
>
>*No changes were made to the fiscal year 2020 peer group used to assist with fiscal year 2021 pay decisions. The peer group consisted of the following 15 companies:*
