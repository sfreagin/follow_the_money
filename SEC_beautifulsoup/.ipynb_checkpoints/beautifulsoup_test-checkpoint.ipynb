### Test using BeautifulSoup to scrape proxy statement

AAPL proxy filed Jan 6, 2021: https://www.sec.gov/Archives/edgar/data/320193/000119312522003583/d222670ddef14a.htm

```python
import requests
import pprintpp
from bs4 import BeautifulSoup
!pip install html5lib
```

```python
# the headers are required or else the request will time out
# get your header from:
# http://httpbin.org/get

URL = 'https://www.sec.gov/Archives/edgar/data/320193/000119312522003583/d222670ddef14a.htm'

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"}
r = requests.get(URL, headers=headers)

soup = BeautifulSoup(r.content, 'html.parser')
```
